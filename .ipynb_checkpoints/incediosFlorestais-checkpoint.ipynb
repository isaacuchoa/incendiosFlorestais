```python
# import libs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Remove Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data

analise = pd.read_csv('db_Incedios.csv')
```

```python
analise01 = analise.rename(columns={
    'year':'Ano',
    'state':'Estado',
    'month':'Mês',
    'number':'Número',
    'date':'Data',
}, inplace=True)

analise.head()
```

```python
# Mapa de calor para verificar campos vazios
sns.heatmap(analise.isnull());
```

```python
# table describe
analise.info()
```

```python
#Número de queimadas por ano
analise.groupby(by=['Ano']).sum()
```

```python
# gráfico das queimadas por ano
plt.figure(figsize=(12,6));

plt.title("Total incêndios no Brasil 1997 - 2017", loc='center', fontsize=14)
sns.lineplot(data=analise, x='Ano', y='Número', estimator='sum', color='#ff5555');

plt.xlabel('Quantidade')
plt.ylabel("Período");
```

```python
analise.groupby(by=['Ano','Mês']).sum().reset_index()

# Gráfico das queimadas por ano e mês
plt.figure(figsize=(12,5))

plt.title("Total de incêdios por Mês")
sns.boxplot (data=analise, x='Mês', y='Número', 
             order=['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']);
plt.ylabel('Número de Incêndios');
```
