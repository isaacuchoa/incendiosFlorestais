```python
# import libs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Remove Warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# load data

file = pd.read_csv('db_Incedios.csv')
file
```

```python
file.rename(columns={
    'year':'Ano',
    'state':'Estado',
    'month':'Mês',
    'number':'Número',
    'date':'Data',
}, inplace=True)

file.head()
```

```python
# Mapa de calor para verificar campos vazios
sns.heatmap(file.isnull());
```

```python
# table describe
file.info()
```

```python
#Número de queimadas por ano
file.groupby(by=['Ano']).sum()
```

```python
# gráfico das queimadas por ano
plt.figure(figsize=(12,6));

plt.title("Total incêndios no Brasil 1997 - 2017", loc='center', fontsize=14)
sns.lineplot(data=file, x='Ano', y='Número', estimator='sum', color='#ff5555');

plt.xlabel('Quantidade')
plt.ylabel("Período");
```
